- 가장 기본

```python
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512) #모델에 입력될 크기
```

```python
images = model.text_to_image("photograpy, Looking at the night sky full of stars, landscape, 8k uhd", batch_size=1)
```

```python
def plot_image(image, figsize=(12.8,7.2),width=1280, height=720):
    plt.figsize=figsize
    plt.imshow(image, extent=[0, width, 0, height])
    plt.axis("off")
    plt.show()
```

```python
plot_image(images[0],figsize=(12.8, 7.2))
```
