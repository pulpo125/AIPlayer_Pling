> ply_meta data

```python
ply_meta = pd.read_json('ply_meta.json')
ply_meta.head(10)
```

> ply_title_embedding data

```python
ply_title_embedding = pd.read_json('ply_title_numodel_emd_demo.json')
ply_title_embedding.head(10)
```
