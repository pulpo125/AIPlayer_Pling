```python
import numpy as np
import pandas as pd
from numpy import dot
from numpy.linalg import norm
import urllib.request
from sentence_transformers import SentenceTransformer
```

```python
# 최초 코드 실행 시 다운로드 시간이 꽤 걸립니다
model = SentenceTransformer('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
```

```python
import json
from pandas import json_normalize
```

```python
path = './train.json'

with open(path, 'r') as f:
    data = json.load(f)
```

```python
df = json_normalize(data)
# # df = df[:5000]
df = df.sample(1000) # 1000개만 뽑아서 해보깅
df
```

```python
df['embedding'] = df.apply(lambda row: model.encode(row.plylst_title), axis = 1) # 인코딩 코드
```

```python
df # embedding column이 추가된 것을 확인할 수 있음
```

```python
# cosine similarity
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
# 가장 유사한 플레이리스트 추출
def return_answer(question):
    embedding = model.encode(question)
    df['score'] = df.apply(lambda x: cos_sim(x['embedding'], embedding), axis=1)
    return df.loc[df['score'].idxmax()] #['plylst_title']
```

```python
return_answer("주먹이 우는 노래")['plylst_title'] # 입력한 제목과 가장 유사한 플레이리스트
```

```python
return_answer("주먹이 우는 노래")
```

```python
df.sort_values(by='score', ascending=False).head(3) # score가 높은 플레이리스트 3개 추출
```

```python
# input?
title = return_answer("주먹이 우는 노래")
tags = title['tags']
song_id = title['songs']

input = song_id + tags
input
```
