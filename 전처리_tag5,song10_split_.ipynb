### 1. tag 사용횟수 기준으로 자르기

```python
train.explode('tags')['tags'].nunique()
```

```python
# tag 합치기
all_tags = []
for tags in train['tags']:
    for tag in tags:
        all_tags.append(tag)

all_tags
```

```python
len(all_tags)
```

```python
pd.Series(all_tags).nunique()
```

```python
# 태그 카운트
tag_counts = pd.Series(all_tags).value_counts()
tag_counts
```

```python
# 5번 이하로 사용된 태그
unique_tags = list(tag_counts[tag_counts <= 5].index)
unique_tags
```

```python
len(unique_tags)
```

```python
# unique_tags를 집합(set)으로 변환
unique_tags = set(unique_tags)
len(unique_tags)
```

```python
# # 확인용

# test = train.iloc[:10, :]
# tmp_tags = ['팝', '기분좋은', '겨울','다시한번', '수상', '발라드']
# test
```

```python
# # 확인용

# #'tags' 열에서 tmp_tags 중 하나라도 들어있는 값을 삭제
# test['tags'] = test['tags'].apply(lambda x: [tag for tag in x if tag not in tmp_tags])

# # tmp_tags가 더 이상 'tags'에 없지 않은 행만 남김
# test = test[test['tags'].apply(lambda x: len(x) > 0)]
# test
```

```python
# tag 사용횟수가 5번 이하인 태그를 제외
train['tags'] = train['tags'].apply(lambda x: [tag for tag in x if tag not in unique_tags])
```

```python
# tags가 비어있지 않은 행만 남김
train_tags = train[train['tags'].apply(lambda x: len(x) > 0)]
```

```python
train_tags = train_tags.reset_index(drop=True)
```

```python
train_tags.head()
```

```python
# tag가 0개인 플레이리스트가 있는지 확인
count = 0

for i in range(len(train_tags)):
    if len(train_tags['tags'][i]) == 0:
        count += 1

print(count)
```

### 2. song 자르기

```python
train_tags.explode('songs')['songs'].nunique()
```

```python
# song_id 합치기
all_songs = []
for songs in train_tags['songs']:
    for song in songs:
        all_songs.append(song)

all_songs
```

```python
len(all_songs)
```

```python
pd.Series(all_songs).nunique()
```

```python
# 태그 카운트
song_counts = pd.Series(all_songs).value_counts()
song_counts
```

```python
# 10번 이하로 사용된 태그
unique_songs = list(song_counts[song_counts <= 10].index)
unique_songs
```

```python
# unique_songs를 집합(set)으로 변환
unique_songs = set(unique_songs)
len(unique_songs)
```

```python
# song 사용횟수가 10번 이하인 song 제외
train_tags['songs'] = train_tags['songs'].apply(lambda x: [song for song in x if song not in unique_songs])
```

```python
# tags가 비어있지 않은 행만 남김
train_tags_songs = train_tags[train_tags['songs'].apply(lambda x: len(x) > 0)]
```

```python
train_tags_songs = train_tags_songs.reset_index(drop=True)
```

```python
# song이 0개인 플레이리스트가 있는지 확인
count = 0
for i in range(len(train_tags_songs)):
    if len(train_tags_songs['songs'][i]) == 0:
        count += 1

print(count)
```

```python
train_tags_songs.head()
```

```python
train_tags_songs.explode('songs')['songs'].nunique()
```

```python
train_tags_songs.info()
```

```python
# tags의 빈 값이 없는지 다시 확인

train_tags_0 = train_tags_songs.explode('tags')['tags'].sort_values()
train_tags_0
```

```python
tags_0 = train_tags_0.iloc[:10].index.to_list()
```

```python
train = train_tags_songs.drop(tags_0)
```

```python
# tags의 빈 값이 없는지 다시 확인

train.explode('tags')['tags'].sort_values()
```

```python
train['songs_cnt'] = train['songs'].apply(len)
train.sort_values('songs_cnt')
```
